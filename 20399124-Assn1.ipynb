```python
# function that replace infinitie values with nan
def replace_infinite_with_nan(df):
    """Replaces infinite values in a dataframe with NaN.

    Args:
    df: The dataframe to be processed.

    Returns:
    A new dataframe with infinite values replaced with NaN.
    """
    # Convert infinite values to NaN.
    df = df.replace([np.inf, -np.inf], np.nan)

    return df
```

```python
# function that fill nan values
def fill_nan_values(df):
    """
    This function fills NaN values in a DataFrame with the mean for numerical columns and with the mode for object columns.

    Args:
    df: The DataFrame to be processed.

    Returns:
    A new DataFrame with the NaN values filled in.
    """
    
    # Get the columns with NaN values.
    nan_columns = df.columns[df.isna().any()]

    # Loop over the NaN columns.
    for column in nan_columns:
        
        # Get the data type of the column.
        data_type = df[column].dtype

        ints = ['int8', 'int16', 'int32', 'int64']
        floats = ['float16', 'float32', 'float64']
        # Fill the NaN values in the column with the appropriate value.
        if data_type in floats:
            df[column].fillna(df[column].mean(), inplace=True)
        elif data_type in ints:
            df[column].fillna(df[column].mean(), inplace=True)
        elif data_type == "object":
            df[column].fillna(df[column].mode()[0], inplace=True)
        else:
            df[column].dropna(inplace=True)
        
    return df
```

```python
# function that drop nan values
def drop_nan_rows(df):
    """
    Drop all rows with NaN values in a Pandas DataFrame.

    Args:
    df: The Pandas DataFrame to drop NaN rows from.

    Returns:
    A new Pandas DataFrame with all NaN rows dropped.
    """

    # Drop all rows with NaN values.
    df = df.dropna(axis=0, how='any')

    # Reset the index of the DataFrame.
    df = df.reset_index(drop=True)

    return df
```

```python
# function that handle outliers using z score
def handle_outliers_zscore(data, threshold=2):
    """
    This function handles outliers using z-score.

    Args:
    data: The data to be processed.
    threshold: The threshold value for z-score.

    Returns:
    A new dataset with outliers removed.
    """

    # Calculate the mean and standard deviation of the data.
    mean = np.mean(data)
    std = np.std(data)

    # Calculate the z-score for each data point.
    z_scores = (data - mean) / std

    # Identify the data points that are outliers.
    outliers = np.where(np.abs(z_scores) > threshold)[0]

    # Remove the outliers from the data.
    new_data = np.delete(data, outliers, axis=0)

    return new_data
```

```python
# function that encoding the categorical function
def encode_categorical_features(df):
    """
    Encodes categorical features as numerical values so that they can be used by machine learning algorithms.

    Args:
    df: The DataFrame to be encoded.

    Returns:
    A DataFrame with the encoded features.
    """
    
    # Identify the categorical features.
    categorical_features = df.select_dtypes(include='object').columns

    # Encode the categorical features using label encoding.
    le = LabelEncoder()
    for feature in categorical_features:
        df[feature] = le.fit_transform(df[feature])
        
    # Return the encoded DataFrame.
    return df
```

```python
# function that normalize the numerical features
def normalize_numerical_features(df):
    """
    Normalizes numerical features so that they are on a comparable scale.

    Args:
    df: The DataFrame to be normalized.

    Returns:
    A DataFrame with the normalized features.
    """
    
    # Identify the numerical features.
    numerical_features = df.select_dtypes(include='number').columns

    # Normalize the numerical features using min-max normalization.
    for feature in numerical_features:
        df[feature] = (df[feature] - df[feature].min()) / (df[feature].max() - df[feature].min())

    # Return the normalized DataFrame.
    return df
```

```python
# function that plotting heatmap (correlation)
def corrvis(df):
    fig, ax = plt.subplots(figsize=(30, 10))
    cor_df = df.corr(method = "pearson").round(2)
    sns.heatmap(cor_df, ax=ax, annot=True)
    
    # Generate annotation labels array (of the same size as the heatmap data)- filling cells you don't want to annotate with an empty string ''
    annot_labels = np.empty_like(cor_df, dtype=str)
    annot_mask = cor_df >= 0.8
    annot_labels[annot_mask] = 'T' 
    # Plot hearmap with the annotations
    fig, ax = plt.subplots(figsize=(30, 10))
    sns.heatmap(cor_df, annot=annot_labels, fmt='')
```

```python
def reduce_memory_usage(df, verbose=True):
    
    
    '''
    This function reduces the memory sizes of dataframe by changing the datatypes of the columns.
    Parameters
    df - DataFrame whose size to be reduced
    verbose - Boolean, to mention the verbose required or not.
    '''
    numerics = ['int8', 'int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    start_mem = df.memory_usage().sum() / 1024**2
    for col in df.columns:
        col_type = df[col].dtypes
        if col_type in numerics:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            else:
                c_prec = df[col].apply(lambda x: np.finfo(x).precision).max()
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max and c_prec == np.finfo(np.float16).precision:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max and c_prec == np.finfo(np.float32).precision:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
    end_mem = df.memory_usage().sum() / 1024**2
    if verbose: print('Mem. usage decreased to {:5.2f} Mb ({:.1f}% reduction)'.format(end_mem, 100 * (start_mem - end_mem) / start_mem))
    return df
```

```python
def calculate_percentage_of_missing_values(df):
    """
    Calculates the percentage of missing values for each column in a Pandas DataFrame.

    Args:
    df: The Pandas DataFrame.

    Returns:
    A new DataFrame with the percentage of missing values for each column.
    """

    # Count the number of missing values in each column.
    missing_values = df.isna().sum()

    # Calculate the percentage of missing values for each column.
    percentage_of_missing_values = missing_values / len(df) * 100

    # Return a new DataFrame with the percentage of missing values for each column.
    return percentage_of_missing_values
```
